# Pearson correlation
a popular method to calculate correlations between two signals is the Pearson correlation coefficient.  It is calculated as:

$$\rho_{X,Y}=\frac{cov(X,Y)}{\sigma_{X}\sigma_{Y}}$$

with $cov(X,Y)$ being the covariate between $X$ and $Y$, and $\sigma$ being the standard deviation.  In particular, for two data sets $\{x_{i}\}, \{y_{i}\}$ it can be calculated as:

$$\rho_{X,Y} = \frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sqrt{\sum_{i=1}^{N}(x_{i}-\bar{x})^2}\sqrt{\sum_{i=1}^{N}(y_{i}-\bar{y})^2}}$$

```python
from scipy.stats import pearsonr
```

```python
corr_pla=pearsonr(front_pla,amy_pla)
corr_oxy=pearsonr(front_oxy,amy_oxy)
print(corr_pla)
print(corr_oxy)
print(np.arctanh(corr_pla[0]),1/np.sqrt(len(amy_pla)-3))
print(np.arctanh(corr_oxy[0]),1/np.sqrt(len(amy_pla)-3))
```

```python
import pymc3 as pm
```

```python
data = [np.vstack((front_oxy,np.roll(amy_oxy,t))).T for t in range(-3,4)]
print(data[1])
```

```python
with pm.Model() as model:
    #priors
    t0=pm.Normal('t0',mu=0.0,sd = 1.0)
    a = -pm.Exponential('a',lam=0.01)
    rho0 = pm.Uniform('rho0',-1.0, 1.0)
    sig1=pm.Uniform('sig1',0.0,10.0)
    sig2=pm.Uniform('sig2',0.0,10.0)
    mu1=pm.Normal('mu1',mu=0.0,sd=sig1)
    mu2=pm.Normal('mu2',mu=0.0,sd=sig2)
    
    #likelyhood function
    distribution1=pm.MvNormal('distribution1',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(-3.0-t0)**2+rho0)*sig1*sig2],[(a*(-3.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[0])
    #likelyhood function
    distribution2=pm.MvNormal('distribution2',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(-2.0-t0)**2+rho0)*sig1*sig2],[(a*(-2.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[1])
    #likelyhood function
    distribution3=pm.MvNormal('distribution3',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(-1.0-t0)**2+rho0)*sig1*sig2],[(a*(-1.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[2])
    #likelyhood function
    distribution4=pm.MvNormal('distribution4',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*t0**2+rho0)*sig1*sig2],[(a*t0**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[3])
    #likelyhood function
    distribution5=pm.MvNormal('distribution5',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(1.0-t0)**2+rho0)*sig1*sig2],[(a*(1.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[4])
    #likelyhood function
    distribution6=pm.MvNormal('distribution6',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(2.0-t0)**2+rho0)*sig1*sig2],[(a*(2.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[5])
    #likelyhood function
    distribution7=pm.MvNormal('distribution7',
                          mu=[mu1,mu2],
                          cov=[[sig1**2,(a*(3.0-t0)**2+rho0)*sig1*sig2],[(a*(3.0-t0)**2+rho0)*sig1*sig2,sig2**2]],
                          observed=data[6])
    
    start = pm.find_MAP()
    trace = pm.sample(10000,start=start)
```

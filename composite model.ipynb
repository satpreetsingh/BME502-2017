```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pymc as pm
```

```python
c=np.concatenate((np.ones(5),np.ones(5)*2,np.ones(5)*3))
```

```python
x = np.array(list(range(5))*3)
print(x,c)
```

```python
y = x*0.5*c + np.random.normal(0,1,15)
```

```python
plt.plot(x[:5],y[:5],"o")
plt.plot(x[5:10],y[5:10],"o")
plt.plot(x[10:15],y[10:15],"o")
```

```python
a = pm.Normal('a',mu=1,tau=1/25)
myy = pm.Normal('myy',mu = x*a*c, tau=1,value=y, observed=True)
```

```python
m=pm.MCMC([a,myy])
m.sample(iter=10000,burn=1000,thin=10)
```

```python
pm.Matplot.plot(m)
```
